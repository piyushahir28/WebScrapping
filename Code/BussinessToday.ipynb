````python
import re
import csv
from time import sleep
from bs4 import BeautifulSoup
from bs4 import * 
import requests
import pandas as pd
````

````python
def get_article(head):
  headline = head.find('h2').a.text
  posted = head.find('span','').text.replace("New Delhi, ","").replace("Mumbai, ","").replace("Delhi,","")

  article = (headline,posted)
  return article
````

````python
def get_the_news(totalPage):
  tPage = totalPage
  articles = []
  
  for page in range(1, tPage):
    template = 'https://www.businesstoday.in/search.jsp?searchword=Reliance_Industries&page={}'
    url = template.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    heads = soup.find_all('fieldset')
  
    for head in heads:
      article = get_article(head)
      articles.append(article)

    sleep(1)

    with open('results.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['Headline', 'Posted'])
        writer.writerows(articles)
              
  return articles
````
