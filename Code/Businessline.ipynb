````python
import re
import csv
from time import sleep
from bs4 import BeautifulSoup
from bs4 import * 
import requests
import pandas as pd
````

````python
def get_article(head):
  headline = head.find('h3').text
  posted = head.find('span','artdate').span.text

  article = (headline,posted)
  return article
````

````python
def get_the_news(totalPage):
  tPage = totalPage
  articles = []
  
  for page in range(1, tPage):
    template = 'https://www.thehindubusinessline.com/search/?order=DESC&page={}&q=Reliance%20Industries&s=markets&sort=publishdate&t=Reliance%20Industries%20Ltd&tid=tag%3Acompanies_bl.ewf%2C2017%3Areliance-industries-ltd'
    url = template.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    heads = soup.find_all('article', 'stock-markets')
  
    for head in heads:
      article = get_article(head)
      articles.append(article)

    sleep(1)

    with open('Businessline.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['Headline', 'Posted'])
        writer.writerows(articles)
              
  return articles
````

````python
articles = get_the_news(20)
````
